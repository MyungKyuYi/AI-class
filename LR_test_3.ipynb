##breast cancer / Decision Tree

트리는 본질적으로 if-else 규칙들의 집합이다.

분류용 결정트리에서 트리는 각 노드에서
“어떤 기준으로 나누면 클래스가 가장 잘 구분될까?”를 찾는다.

단점

과적합(overfitting) 이 잘 된다

데이터가 조금만 바뀌어도 구조가 크게 달라질 수 있다

일반화 성능이 불안정할 수 있다

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

path = '/content/drive/MyDrive/Colab Notebooks/인공지능개론/breast_cancer (1).csv'
df = pd.read_csv(path)

y=df['label']
y.value_counts() #각 클래스가 몇 개 있는지 확인하는 코드 불균형한지 확인한다
X=df.drop('label',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)

from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0) #결정트리 모델 생성
clf_dt.fit(X_train, y_train) #학습

pred_dt = clf_dt.predict(X_test) #예측

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))

#결과
# accuracy_score: 전체 중 맞춘 비율
# [[실제 0을 0으로 예측, 실제 0을 1로 예측],
#  [실제 1을 0으로 예측, 실제 1을 1로 예측]]
```
