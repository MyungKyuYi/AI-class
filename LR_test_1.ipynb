##breast cancer / Logistic Regression

```python
import pandas as pd
import numpy as np
path = '/content/drive/MyDrive/Colab Notebooks/인공지능개론/breast_cancer (1).csv'
df = pd.read_csv(path)
```

```python
y=df['label']
y.value_counts()
X=df.drop('label',axis=1)
#axis는 어느 축을 기준으로  1은 열(column) 기준, 0은 행(row) 기준
# y는 정답 값, x는 입력 특성
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
# x, y는 위치인자, test_size=0.25, random_state=0 → 키워드 인자(keyword arguments)이름을 지정해서 전달함
# 이떄 train_test_split()은 4개의 값을 반환
# 데이터를 학습용 75%, 테스트용 25%로 나눈다.
# random_state=0은 같은 분할을 재현하기 위한 설정이다. 즉 랜덤결과 고정

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)
pred_lr = clf_lr.predict(X_test)
# 로지스틱 회귀 모델을 학습
# 테스트 데이터에 대해 예측 수행

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))

# accuracy_score: 전체 정확도
# confusion_matrix: 어떤 클래스를 얼마나 맞추고 틀렸는지 보여줌

#결과

# 정확도: 0.9720
# 실제 0인 것 중 52개 맞추고 1개 틀림
# 실제 1인 것 중 87개 맞추고 3개 틀림
```
