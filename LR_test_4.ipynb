##breast cancer / RandomForest

결정트리를 여러 개 만든 뒤 합치는 앙상블(ensemble) 모델

분류일 때 / 여러트리가 예측한 클래스를 다수결 투표

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

path = '/content/drive/MyDrive/Colab Notebooks/인공지능개론/breast_cancer (1).csv'
df = pd.read_csv(path)

y=df['label']
y.value_counts()
X=df.drop('label',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)

from sklearn.ensemble import RandomForestClassifier

print ("\n--- Radom Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)

pred = rf_clf.predict(X_test)
# 테스트 샘플 하나에 대해 여러 트리가 각각 예측한다.
# 분류 문제에서는 최종적으로 다수결 투표(majority voting) 를 한다.

print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```
